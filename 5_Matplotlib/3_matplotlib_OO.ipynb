# Matplotlib OO method

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(-4, 4, 32)
y = x**2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_title('Title')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_title('Larger Title')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')

axes2 = fig.add_axes([0.3, 0.5, 0.5, 0.3])
axes2.plot(x, x)
axes2.set_title('Smaller Title')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
```

```python
fig = plt.figure(figsize=(8,8))
axes1 = fig.add_axes([0,0.1,1,0.3])
axes1.plot(x, y)
axes1.set_title('axes 1')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2 = fig.add_axes([0,0.5,1,0.3])
axes2.plot(x,x)
axes2.set_title('axes 2')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
```

```python
fig.savefig('my_pic.png', dpi=100)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, label='y= x**2')
ax.plot(x, x, label='y= x')
#ax.legend(loc=(0.05,0.3))
ax.legend(loc=0)
```

```python
x = np.linspace(0, 10, 10)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
#ax.plot(x, x**2, color='red')
ax.plot(x, x**2, color='blue', lw=1, alpha=0.5, ls='--', marker='o', markersize=20,
        markerfacecolor='yellow', markeredgecolor='green', markeredgewidth=5)
ax.set_xlim([0,2])
ax.set_ylim([0,5])
```
