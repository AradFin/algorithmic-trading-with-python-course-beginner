```python
df1['A'].plot.hist(bins=30)
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C'] * 100, alpha=0.3)
```

```python
df1.plot.hexbin(x='A', y='B', gridsize=10, cmap='coolwarm')
```
