# Matplotlib : functional method

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# plt.show()
```

```python
import numpy as np
x = np.linspace(-4, 4, 32)
y = x ** 2
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('My Title')
#plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```
